# Categorical Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
# Categorical data is nothing but strings 
```

```python
# ploting categorical column with numerical column
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
# by default it will perform avg on each categorical value 
```

```python
import numpy as np
```

```python
# We can change the aggrigarte function and group the similar data 
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
# it will group by standard deviation
```

```python
# Now lets learn about count plot 
# it is same as barplot in this estimator explicitly count the number of 
# occurence
# Thats why we give only x value
```

```python
sns.countplot(x='sex',data = tips)
```

```python
# Lets understand about box plots
# Box plot actually divide the dataset according to median it further 
# divide half of the data and find 25th percentile and 75th percentile
# last it will add the boxes together to represent data in percentile
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
# dotted lines are actually outliers.....
```

```python
# Now we can add how the other data we can add
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# Now violinplots
# it is same as box plot 
# it will give a lot of information about every data point
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# we can use here hue and split the two values for make it more 
# informative
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
# Now we will learn simple strip plot
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
# swarmplots
# it is just combination of stripplot and violon plot
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# we will now make the combination of violin and and swarmplot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
# Now we can impliment all the above graphs through factor plots
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

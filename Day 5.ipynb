# Pandas - Missing Data

```python
import numpy as np
import pandas as pd
```

```python
a = {'A':[1,2,np.nan],'B':[4,np.nan,np.nan],'C':[6,8,7]}
df1 = pd.DataFrame(a)
df1
```

```python
df1.dropna()
```

```python
df1.dropna(axis = 1)
```

```python
# if ypu want atleast 2 values which are non-na then you can use threshhold
# in a row
```

```python
df1.dropna(thresh = 2)
```

```python
# To fill the na with values we use fillna
```

```python
df1.fillna(value ="Fill Value")
```

```python
df1.fillna(value = df1['A'].mean())
```

# Groupby 

```python
# Create dataframe
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df2 = pd.DataFrame(data)
df2
```

```python
bycomp = df2.groupby('Company')
```

```python
bycomp.mean()
```

```python
bycomp.sum()
```

```python
bycomp.std()
```

```python
# All in a single line 
```

```python
df2.groupby('Company').mean()
```

```python
df2.groupby('Company').describe()
```

```python
df2.groupby('Company').describe().transpose()['FB']
```

#  Merging Joining and Concatinating

```python
dframe1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
dframe2= pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
```

```python
dframe3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
dframe1
```

```python
dframe2
```

```python
# For concatination dimesions must be same of every dataframe
pd.concat([dframe1,dframe2,dframe3])
```

```python
# By default concat method combines the rows
```

```python
pd.concat([dframe1,dframe2,dframe3],axis = 1)
```

```python
# Merging
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})    
```

```python
pd.merge(left,right, how = 'inner', on = 'key')
```

```python
# Merging 2 dataframes with 2 keys  and types of joins
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(left,right,how='inner',on=['key1','key2'])
```

```python
pd.merge(left,right,how = 'left',on=['key1','key2'])
```

```python
pd.merge(left,right,how = 'right',on=['key1','key2'])
```

```python
pd.merge(left,right,how = 'outer',on=['key1','key2'])
```

```python
# joining 
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left.join(right)  # by default inner join
```

```python
left.join(right,how = 'outer')
```

```python
left.join(right,how = 'left')
```

```python
left.join(right,how = 'right')
```

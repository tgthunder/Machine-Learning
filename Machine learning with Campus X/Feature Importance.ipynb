# Feature importance technique in d-tree and random forest

```python
import numpy as np 
import pandas as pd
```

```python
from sklearn.datasets import make_classification,load_digits
```

```python
X,y = make_classification(n_samples=40,n_features=2,n_redundant=0,n_informative=2,n_classes=2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X,y)
```

```python
from sklearn.tree import plot_tree
```

```python
import matplotlib.pyplot as plt
```

```python
plot_tree(clf)
plt.show()
```

```python
clf.feature_importances_
# we can clearly say that 0th feature is more important as compared to 1st
```

**Feature importance in Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=3)
```

```python
rf.fit(X,y)
```

```python
rf.feature_importances_
```

- In case of random forest we calculate avg of each feature imporatnce and then resultant would be the feature importance of the entire random forest

```python
# checking the feature importance of each d-tree in random forest
print(rf.estimators_[0].feature_importances_) # feature importance of 1st dtree
print(rf.estimators_[1].feature_importances_) # feature importance of 2st dtree
print(rf.estimators_[2].feature_importances_) # feature importance of 3st dtree
```

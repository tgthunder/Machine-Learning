```python
import numpy as np 
import pandas as pd
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100,n_features=5,n_redundant=0,n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X,columns=["A","B","C",'D',"E"])
df["Target"] = y
df.head()
```

```python
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
```

```python
bag = BaggingClassifier(max_features=2) # means column sampling
rf = RandomForestClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,:5],df.iloc[:,-1])
rf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
from sklearn.tree import plot_tree
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,12))
plot_tree(bag.estimators_[1])
plt.show()
```

```python
plt.figure(figsize=(10,12))
plot_tree(rf.estimators_[1])
plt.show()
```

**Above code shows that there is difference between bagging with dtree and random forest during column sampling**

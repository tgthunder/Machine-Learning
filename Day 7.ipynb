# Basics Of Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
# it will going to show you plots in jupyter notebook
```

```python
# Data
x = np.linspace(1,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
# There are 2 methods to show the graphs
# 1.Functional
# 2.Object Oriented method
```

```python
# 1.Functional
plt.plot(x,y,'g--')
# This 'g-- is used to determine the color and linstyle'
plt.show()
```

```python
# Adding lables and title
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Tilte')
plt.show()
```

```python
# Ploting multiple graplhs on same canvas
# We will create  two sub graphs using different styles
# Last argument of subplot is numbering of plot
plt.subplot(1,2,1)  
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.show()
```

```python
# Object Oriented method to plot the graphs
# we have to plot all graphs on axes
```

```python
# For adding axes to graph we have to mention 4 arguments Left,Bottom,Width
# Height
fig = plt.figure()
#It will going to create a blanck canvas
axes = fig.add_axes([0.1,1,0.8,1])
axes.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,1])
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
axes.plot(x,y,'r--')
plt.show()
```

```python
# We have to take axes values in range 0-1
# you can create any type of empty graph format
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot()
axes2.plot()
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'g.')
axes1.set_title("Larger Plot")
axes2.plot(y,x,'r-.')
axes2.set_title("Samller plot")
plt.show()
```

# More On Object Oriented Method

```python
# subgraphs using oo
```

```python

fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()

#tight_layout will remove the overlaps for clear presentation
```

```python
# Axes is just a numpy array which we have created manually 
# we can iterate them 
fig,axes=plt.subplots(nrows=1,ncols=2)
for current_axes in axes:
    current_axes.plot(x,y)
plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("1st Graph")
axes[1].plot(y,x,'r--')
axes[1].set_title("2nd Graph")
plt.tight_layout()
# must seperate the layouts
plt.show()
```

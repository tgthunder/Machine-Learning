# Seaborn

```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Dist plot --> used for univariable distribution(means one value)
# it will load unique vale in the distribution 
# It is the type of histogram
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
# kde is another type of plot
# bins are used to set the number of bins in your dataset depending upon data
```

```python
# Joint plots 
# this type of plots are used to compare two distribution plots
# To find some relation between two colums in a dataset
# By default it is scatter plot
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
# The above graph says that if total_bill increses then tip will increse
# that is total_bill is directly pointing to tip
```

```python
# We can determine the above relationship using hexagon plot 
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
# Strong relationship shown by dark color
```

```python
# Now we will drop regression line on it.
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
# Now we will use kde to see the maximum density
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde',shade = True,color='yellow')
```

```python
# Pair plots--> They are used to pair all the numerical value plots 
# in the dataset
```

```python
sns.pairplot(tips,hue='sex')
```

```python
# Now lets discuss about some parameters for pair-plots
# 1.hue-->it is used to plot non-numerical data on plots
# 2.pallete=it is used to change the color
```

```python
sns.pairplot(tips,hue='sex',palette='cool')

#palette-coolwarm,cubehelix,spectral,icefire
```

```python
# rugplots
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
# Actually compare rugplot and distplot. In this we actually we add all the 
# lines of rugplots and add it in distplots
```

```python
# kde --> This type of graph used to represent probability density function
# How it will create?-->
# we just actually add a kernal centered about each data point and finally 
# sum up all kernals resulting the kde
```

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.head()
```

```python
# For univarient data
sns.kdeplot(cars['horsepower'],shade=True)
```

```python
# bandwidth
sns.kdeplot(cars['horsepower'],bw=100,shade=True)
```

```python
#Bivarinte data
# We can find relationship between two datasets using kde plots
```

```python
cars.head()
```

```python
sns.kdeplot(data=cars,x='horsepower',y='mpg',n_levels=20)
# n_levels are used to increase your lines in the data
```

```python
sns.scatterplot(data=cars,x='horsepower',y='mpg',hue='origin')
```

```python
cars['origin'].unique()
```

```python
cars.plot(x='mpg',y='horsepower',kind='scatter',color='yellow')
```

```python
sns.heatmap(cars.corr(),annot=True)
```

```python
sns.factorplot(cars['weight'] , cars['horsepower'] , kind='scatterplot')
```

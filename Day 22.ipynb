# K Means Clustering

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,
                 random_state=101)
```

```python
print(type(data))
```

```python
#now amke a scatter plot to see the data
```

```python
sns.set_style('darkgrid')
```

```python
plt.scatter(data[0][:,0],data[0][:,1])
```

```python
# now see the second element also and visualize it with seperatly
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
# Now we will creste kmeans model and fit our data
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig,(ax1,ax2)= plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.set_title("K means")
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title("K means oroginal")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
plt.tight_layout
```

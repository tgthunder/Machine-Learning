# Grids

```python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
iris=sns.load_dataset('iris')
iris.head()
```

```python
#Now we will learn about grids 
# Grids are just empty graphs of all numrical value pair graphs
```

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
# It will return all the graphs in scatter format
```

```python
# Now I want to give specification for plots according to me
```

```python
g = sns.PairGrid(iris)
# g.map_offdiag(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# Now we will understand about FacetGrid--> It is used to plot categorical 
# data
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.set_style('white')
```

```python
sns.FacetGrid(penguins);
```

```python
g = sns.FacetGrid(penguins,col='island')
g.map(sns.histplot,"flipper_length_mm");
```

```python
# Another method to create facetgrid which allows us to work with 
#Both the axis
```

```python
g = sns.FacetGrid(penguins,col='island')
g.map_dataframe(sns.scatterplot,x='bill_depth_mm',y='bill_length_mm')
```

```python
g=sns.FacetGrid(penguins,col='island',row='species')
g.map_dataframe(sns.scatterplot,x='bill_length_mm',y='bill_depth_mm');
```

# Regression Plots

```python
# linear Model plots
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips);
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws = {'s':100});
```

```python
# We can further divide the data using girds or facetgrids instead of
# hue 
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time');
```

```python
# Now still we want to plot lot of info
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex');
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex');
```

```python
# The above data is little bit difficult to see clearly 
# we mention aspect ratio--> between height and width
# and size == between graph and data
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
          aspect=0.6,size=6);
```

# Styles And Colors

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.countplot(x='sex',data=tips)
```

```python
# we can remove spines from graph
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

```python
# We can size the canvas using matplotlib in seaborn
```

```python
plt.figure(figsize=(10,18))
sns.countplot(x='sex',data=tips)
```

```python
# sns.set_context(context='poster',font_scale=2)
sns.countplot(x='sex',data=tips)
```

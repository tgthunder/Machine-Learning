# Plot Appearence

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(1,5,10)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r--o')
plt.show()
```

```python
fig = plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,'g-->')
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color = 'purple',lw=10,alpha=0.5)
# alpha is used for transperency
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color = 'purple',lw=1,ls='-',marker='o',markersize=15,
         markerfacecolor='y',markeredgewidth=3,markeredgecolor='g')
plt.show()
```

# Plot Range

```python
# retriving graph from a specific range
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='orange',lw=2,ls='--')
axes.set_xlim([1,2])
axes.set_ylim([1,2])
# Set the limit in upper bound and lower bound

plt.show()
```

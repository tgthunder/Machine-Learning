```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# jointplots
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde',shade=True)
```

```python
tips.head()
```

```python
sns.pairplot(tips,hue='sex',palette='cubehelix')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'],shade=True)
```

```python
cars=sns.load_dataset('mpg').dropna()
```

```python
cars.head()
```

```python
sns.kdeplot(x='horsepower',y='mpg',data=cars,n_levels=20)
```

```python
sns.displot(tips['total_bill'],color='red',bins=20)
```

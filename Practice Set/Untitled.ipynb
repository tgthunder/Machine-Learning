```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
x = np.random.randn(10)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'mo--',lw =1,markersize=10)
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Delhi Temperature")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'m')

plt.subplot(2,2,2)
plt.plot(y,x,'r')
plt.subplot(2,2,3)
plt.plot(x,y,'g')

plt.subplot(2,2,4)
plt.plot(y,x,'c')
plt.show()
```

```python
# OO Method for plotting the graphs
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.set_xlabel("Days")
ax.set_ylabel("Temperature")
ax.set_title("Delhi Temperature")
ax.plot(x,y,'ro--',lw=1,markersize=10)
plt.show()
```

```python
# subplots------>
```

```python
fig,ax = plt.subplots(nrows = 1,ncols = 2)
ax[0].plot(x,y)
ax[0].set_title("First Graph")
ax[1].plot(y,x,'r--')
ax[1].set_title("2nd Graph")
plt.tight_layout()
plt.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,4))
axes[0].plot(x,y,'ro--')
axes[1].plot(y,x,'m')
plt.tight_layout()
plt.show()
```

```python
fig.savefig("MYSubploting.png",dpi=200)
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([1,1,1.5,1.5])
ax.plot(x,y,color = 'orange',linestyle = '-',linewidth = 4,marker = 'o',
       markersize = 20 , markerfacecolor = 'yellow', markeredgewidth = 3,
       markeredgecolor = 'green')
ax.set_xlim([-1,0])
ax.set_ylim([0,1])
plt.show()
```

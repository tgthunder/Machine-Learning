# Decision Trees And Random Forests

```python
import numpy as np 
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('kyphosis.csv')
```

```python
df.head()
```

```python
'''This dataset represnts after the surgery kyphosis disease present in the 
patient or not'''
```

```python
df.info()
```

```python
sns.pairplot(df,hue='Kyphosis')
```

```python
# Now split your data
```

```python
X = df.drop('Kyphosis',axis=1)
```

```python
y = df['Kyphosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# now lets go and start of signle decision tree
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predict = dtree.predict(X_test)
```

```python
y_test = np.array(y_test)
y_test
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predict))
print('\n')
print(classification_report(y_test,predict))
```

```python
# We will make the tree
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dtree,filled=True)
plt.tight_layout()
```

```python
# Now lets see how these results are compared to random forest model
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,rfc_pred))
print('\n')
print(classification_report(y_test,rfc_pred))
```
